```python
import pandas as pd
import glob
```

```python
def apresentar_primeiras_e_ultimas_com_reticencias(data):
    # Obter as 3 primeiras linhas
    primeiras_linhas = data.head(3)
    
    # Obter as 3 últimas linhas
    ultimas_linhas = data.tail(3)
    
    # Criar uma linha de reticências
    reticencias = pd.DataFrame({col: ['...'] for col in data.columns}, index=['...'])
    
    # Concatenar as partes
    apresentacao = pd.concat([primeiras_linhas, reticencias, ultimas_linhas])
    
    return apresentacao
```

```python
data=pd.read_excel("Prices_Excel/processed_data/CL/products_company/CL_No_Client.xlsx") 
data
```

```python
all_companies= data['Apple iPhone 13 128GB']
all_companies=all_companies.tolist()
all_companies
#type(all_companies)
```

```python
all_products=data.columns

all_products
```

```python


# Criando um dicionário para armazenar os produtos vendidos por cada empresa
produtos_por_empresa = {companie: [] for companie in all_companies}


for companie in all_companies:
    for produto in all_products:
        if companie in data[produto].values:
            produtos_por_empresa[companie].append(produto)

# Criando um novo DataFrame vazio para armazenar os dados reorganizados
df_companies_products = pd.DataFrame(index=data.columns, columns=all_companies)  # Definindo as colunas como empresas do novo DataFrame

# Preenchendo o novo DataFrame com os produtos vendidos por cada empresa
for companie, produtos in produtos_por_empresa.items():
    df_companies_products.loc[produtos, companie] = produtos

df_companies_products.reset_index(drop=True, inplace=True)
apresentar_primeiras_e_ultimas_com_reticencias(df_companies_products)
    
```

```python

def companies_produduct(path):
    data=pd.read_excel(path)
    all_companies= data['Apple iPhone 13 128GB']
    all_companies=all_companies.tolist()
    all_products=data.columns
    produtos_por_empresa = {companie: [] for companie in all_companies}
    for companie in all_companies:
        for produto in all_products:
            if companie in data[produto].values:
                produtos_por_empresa[companie].append(produto)
    df_companies_products = pd.DataFrame(index=data.columns, columns=all_companies) 


    for companie, produtos in produtos_por_empresa.items():
        df_companies_products.loc[produtos, companie] = produtos

    df_companies_products.reset_index(drop=True, inplace=True)
    return df_companies_products
    

    
```

```python
path = "Prices_Excel/processed_data/CL/products_company"  
list_path_cl = glob.glob(path + "/*.xlsx")
list_path_cl
```

```python
 
list_save_name = [item.replace('.xlsx', '_companies_product.xlsx') for item in list_path_cl]


list_save_name
```

```python
for paths in list_path_cl:
    for save_name in list_save_name:
        df_companies_products=companies_produduct(paths)
        df_companies_products.to_excel(save_name, index=False)

    #df_companies_products
```
